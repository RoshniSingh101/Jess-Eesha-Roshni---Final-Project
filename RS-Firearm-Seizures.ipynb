# Observations

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# first graph of total count of firearm seizures
# load data
data = pd.read_csv("Firearm_Seizures.csv", index_col="_id") # use the column named _id as the row index

# September 2022 data sorted by total number of incidents in descending order
df = pd.DataFrame(data).loc[5281:5340].sort_values(by="total_count", ascending=False)
df = df.dropna() # drop null data

# ignore these neighborhoods because population.csv does not have them
ignored_neighborhoods = ["Mount Washington", "Hazelwood", "South Shore"]

# Filter the DataFrame to exclude the specified neighborhoods
df = df[~df["neighborhood"].astype(str).isin(ignored_neighborhoods)]

# change size of graph to fit x-axis labels
plt.figure(figsize=(70, 20))

# plot the data using bar() method 
plt.bar(df["neighborhood"].astype(str), df["total_count"]) # x-axis = neighborhoods & y-axis = total count of firearm incidents
plt.title("Total Firearm Seizures in Allegheny County Neighborhoods", fontsize=14) 
plt.xlabel("Neighborhoods", fontsize=12) 
plt.ylabel("Number of Firearm Seizures", fontsize=12) 
plt.show()
```

The neighborhoods with the least number of firearm seizures are East Hills, Knoxville, East Allegheny, Homewood North, Lower Lawrenceville, Crafton Heights, Fineview, California-Kirkbride, Homewood West, Allentown, Homewood South, Larimer, Bluff, Central Business District, North Shore, Carrick, Highland Park, Bon Air, Brighton Heights and Garfield with one incident present for each neighborhood, making them have the potential to be one of the safest overall out of all of the neighborhoods listed in Allegheny County. Firearm seizures include pistols, rifles, revolvers, shotguns, and other firearms.
